### Lab4: Data Visualization 

Ever wonder how much to tip your waiter? One dedicated waiter meticulously recorded information about each tip he received over a few months while working at a restaurant. In total, he documented 244 tips. Now, will use the visualization techniques that we learned to get some insights from this dataset. The data is provided for you in the tips.csv file

```python
import pandas as pd
import seaborn as sns

tips_data = pd.read_csv('tips.csv')
tips_data.head()
```

#### Bar Charts

Q1: Create a bar chart showing the average tip amount for each day of the week. What trend do you notice about tips across different days?



```python
sns.barplot(x="day", y="tip", data=tips_data)
```

Q2: Create a bar chart to compare the average total bill amount by gender. Is there a difference in average spending between male and female customers?

```python
sns.barplot(x="sex", y="total_bill", data=tips_data, ci=None)
```

#### Stacked Bar Charts

Q3: Draw a stacked bar chart that shows the number of non-smokers and smokers for each day. What can you infer about the smoking habits of customers on different days?

```python
sns.countplot(x="day", hue="smoker", data=tips_data)
```

Q4: Plot a stacked bar chart showing the meal counts (lunch vs. dinner) by day. Which meal is more popular on different days?

```python
sns.countplot(x="day", hue="time", data=tips_data)
```

##### Histograms

Q5: Plot a histogram of the total bill amounts. Describe the distribution shape and identify any noticeable patterns.


```python
sns.histplot(tips_data['total_bill'], kde=True)
```

Q6: Create a histogram of tip amounts. Are there any common tip values that customers tend to give?

```python
sns.histplot(tips_data['tip'], kde=True)
```

##### Scatter Plots

Q7: Create a scatter plot of total bill vs. tip amount, with points colored by gender. Do you notice any differences in tipping patterns between male and female customers?


```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips_data)
```

Q8: Plot a scatter plot showing the total bill vs. tip amount, with points colored by whether the customer is a smoker. Is there any visible trend between smokers and non-smokers regarding tipping?

```python
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips_data)
```

#### Boxplots

Q9: Create a boxplot showing the distribution of total bill amounts by day. Which day has the widest range of bill amounts?


```python
sns.boxplot(x="day", y="total_bill", data=tips_data)
```

Q10: Draw a boxplot comparing tip amounts by gender. Do both genders show similar tipping ranges, or is there a noticeable difference?

```python
sns.boxplot(x="sex", y="tip", data=tips_data)
```

#### FacetGrid Graphs

Q11: Use a FacetGrid to plot histograms of tip amounts for each day. How do the tip distributions differ by day?

```python
facet_grid_1 = sns.FacetGrid(tips_data, col="day")
facet_grid_1.map(sns.histplot, "tip", kde=True)
```

Q12: Create a FacetGrid scatter plot showing total bill vs. tip for each day. Are there any days where higher tips are consistently given?

```python
facet_grid_2 = sns.FacetGrid(tips_data, col="day")
facet_grid_2.map(sns.scatterplot, "total_bill", "tip")
```

#### Heatmap Graph

Q13: Create a heatmap to show the correlation between total bill, tip, and size. Which variables are most strongly correlated with each other?

```python
correlation_matrix = tips_data[['total_bill', 'tip', 'size']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
```

#### Pairplot Graph

Q14: Generate a pairplot to examine the relationships among total bill, tip, and size. What patterns or relationships do you observe across these variables?

```python
sns.pairplot(tips_data[['total_bill', 'tip', 'size']])
```
