##### Exercise 1 (iris dataset):

iris dataset has 50 samples of 3 different species of iris (150 samples total). Each sample has four measurements: sepal length, sepal width, petal length, petal width. In this exercise, we will build a machine learning model that can predict the iris species based on given measurements. 



##### Step1: Getting the data ready
In this step, you need to do the following:


1. Load the dataset
2. Seperate the featuers and the target
3. Split the dataset to training and testing sets

```python
# Write your answer for step1 here

import pandas as pd
iris=pd.read_csv("iris.csv")
iris.head()
```

```python
# Write your answer for step1 here

from sklearn.model_selection import train_test_split
iris["variety"].unique()

x=iris.drop("variety", axis=1)

y= iris["variety"]
y.head()

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20)
print(x_train.shape)
print(x_test.shape)
```

##### Step2: Choosing the model
The problem is clearly a classification problem. This is because the data is labled and we know the target column. Additionally, the target is categorical (not numerical). So, some of the possible algorithms that can be used are:

1. KNN (K nearest neighbors) classifier
2. Logistic Regression
3. Support Vector Classifier

##### Step3: Fitting the model

As we saw in the lecture before, fitting the model will be done in few lines of code as the following:

1. Import the class you plan to use
2. Make an instance of the model that you imported
3. Fit the model with data (aka "model training")
4. Predict the response for a new observation (optional)

```python
# Write your answer for step3 here
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train, y_train)
```

##### Step4: Evaluating the model

At this stage, we have the model trained already. However, we still need to evalute how good is our model.This step consists of two steps as the following:

1. Making predictions on the testing set
2. Comparing actual response values (y_test) with predicted response values


```python
# Write your answer for step4 here

y_pred = knn.predict(x_test)

from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred))
```

##### Step5: Expirement to improve

Each model has different hyperparameters that can be tuned to improve the model. Although there is no clear guidlines to idetify what are the setting that will lead to highest accuracy, different values can be tested until we are satisifed with the model. For example, n_neighbors parameter can be changed in odrer to improve the KNeighborsClassifier model performance.

```python
# Write your answer for step5 here
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)

from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred))
```

#### Exercise 2: (red wine quality dataset)

In this exercise, we will work on wine quality dataset. The dataset includes data for Portuguese red and white "Vinho Verde" wines, with variables based on physicochemical properties and sensory quality scores. This dataset can be used for classification or regression tasks, though the classes are imbalanced.

##### Step1: Getting the data ready
In this step, you need to do the following:


1. Load the dataset
2. Seperate the featuers and the target
3. Split the dataset to training and testing sets

```python
# Write your answer for step1 here

wine = pd.read_csv("winequality-red.csv")
wine.head()


X = wine.drop("quality", axis=1)  # Features
y = (wine["quality"] > 5).astype(int)  

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

wine["quality"].value_counts()
```

##### Step2: Choosing the model

What are the models that you think are appropriate for this exercise?


```python
# Write your answer for step 2 here
# Logistic Regression, Reason: works well for binary classification when the relationship between features and the target is mostly linear.
```

##### Step3: Fitting the model

You need to do the following:

1. Import the class you plan to use (use KNN and Logistic Regression)
2. Make an instance of the model that you imported
3. Fit the model with data (aka "model training")


```python
# Write your answer for step3 here

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver="liblinear")

logreg.fit(x_train, y_train)
```

##### Step4: Evaluating the model

At this stage, the following tasks are needed:

1. Making predictions on the testing set
2. Comparing actual response values (y_test) with predicted response values
3. Which model had the heighest accuracy? Why do you think that we did not have a high accuracy for our models?

```python
# Write your answer for step4 here


y_pred = logreg.predict(x_test)

from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred))
```
